```python
# 13.Evening discount

time=int(input("Enter current time:"))
price=float(input("Enter the product price:"))
if time>=18 and time<=21:
    discount=price*0.20
    new_price=price-discount
    print("Final price is:",new_price)
else:
    print("Sorry! No discount")
    print("Final price is:",price)
```

```python
# 17.Bmi Calculator

weight=float(input("Enter your weight:"))
height=float(input("Enter your height:"))
BMI=weight/(height*height)
if BMI<18.5:
    print("Your BMI is:",BMI)
    print("Underweight")
elif BMI>=18.5 and BMI<=24.9:
    print("Your BMI is:",BMI)
    print("Normal")
else:
    print("Your BMI is:",BMI)
    print("Overweight")
```

```python
# 19.Flight fare check

age=int(input("Enter your age:"))
fare=int(input("Enter the fare amount:"))
if age<12 or age>60:
    dis=fare*0.50
    fin_price=fare-dis
    print("Final price for your fare is:",fin_price)
else:
    print("Your fare costs",fare)
```
