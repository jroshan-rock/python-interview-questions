```python
days=int(input("Enter the number of overdays:"))
if days<=5:
    print("Rs.2 per day")
    fine=days*2
    print("Total fine:",fine)
elif 6<=days<=10:
    print("Rs.5 per day")
    fine=days*5
    print("Total fine:",fine)
else:
    print("Rs.10 per day")
    fine=days*10
    print("Total fine:",fine)
```

```python
print("Choose class type :1.Economy class 2.Business class 3.First class")
clas=input("Enter your class type:").lower()
weight=int(input("Enter the weight of luggage:"))
if clas=='economy':
    limit=20
    rate=200
elif clas=='business':
    limit=30
    rate=150
elif clas=='first':
    limit=40
    rate=100
else:
    print("Invalid class type")
    exit()

if weight>limit:
    excess_weight= weight-limit
    excess_fee=excess_weight*rate
    print("Excess weight:",excess_weight)
    print("Excess charge:",excess_fee)
else:
    print("No excess charge")
```

```python
print("Customer type: Regular, Member, VIP")
print("Room type: Deluxe,Standard")

customer=input("Enter customer type:").lower()
room=input("Enter room type:").lower()
price=int(input("Enter room price:"))

if customer=='regular':
    print("You got 10% off")
    dis=price*0.10
    fin=price-dis
    print("Final price is:",fin)
elif customer=='member':
    if room=='deluxe':
        print("You got 20% off on Deluxe room")
        dis=price*0.20
        fin=price-dis
        print("Final price is:",fin)
    elif room=='standard':
        print("You got 15% off on standard room")
        dis=price*0.15
        fin=price-dis
        print("Final price is:",fin)
elif customer=='vip':
    print("You got 30% off")
    dis=price*0.30
    fin=price-dis
    print("Final price is:",fin)
```

```python
print("Choose the time according to given guidelines:")
print("Daytime(6 AM to 10 PM)")
print("Nighttime(10 PM to 6 AM)")
time=input("Enter the time:").lower()
dis=int(input("Enter the distance:"))
if time=='daytime':
    if dis>20:
        fare=dis*15
        discount=fare*0.10
        final=fare-discount
        print("Your final amount is:",final)
    elif dis<=20:
        fare=dis*15
        print("Your final amount is:",fare)
elif time=='nighttime':
    if dis>20:
        fare=dis*20
        discount=fare*0.10
        final=fare-discount
        print("Your final amount is:",final)
    elif dis<=20:
        fare=dis*20
        print("Your final amount is:",fare)
    
```
